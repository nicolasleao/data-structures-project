```julia
using Plots 

x = Int64[]
y = Int64[]
open("melhor_lista") do file
    i = 1
    for line in eachline(file)
        arr = split(line)
        append!(x, i)
        append!(y, parse(Int64, arr[3]))
        i += 1
    end
    plot(x, y, title="Fila sem heap - melhor caso.", label="Número de comparações")
end
```
